```python
import re # regular expression
import pandas as pd
```

```python
f= open('D:\WhatsApp_Chat_Analysis\WhatsApp Chat with Sketchy Hookers .ORG.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
data = data.replace('\u202f', '')
print(data)
```

```python
pattern = r"\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}[\s\u202f]*[ap]m\s-\s"
```

```python
messages= re.split(pattern,data)[1:] ## since getting an empty string 1st row , to remove it used[1:]
messages
```

```python
dates= re.findall(pattern,data) # to extract date
dates
```

```python
df = pd.DataFrame({'user_message': messages,'message_date': dates})
```

```python
df
```

```python
df['message_date']=pd.to_datetime(df['message_date'], format="%d/%m/%y, %I:%M%p - ", dayfirst=True)
```
