```python
import re # regular expression
import pandas as pd
```

```python
f= open("D:\WhatsApp_Chat_Analysis\WhatsApp Chat with DOPE'27 Class A.txt",'r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
data = data.replace('\u202f', '')
print(data)
```

```python
pattern = r"\d{2}/\d{2}/\d{2},\s\d{1,2}:\d{2}[\s\u202f]*[ap]m\s-\s"
```

```python
messages= re.split(pattern,data)[1:] ## since getting an empty string 1st row , to remove it used[1:]
messages
```

```python
dates= re.findall(pattern,data) # to extract date
dates
```

```python
df = pd.DataFrame({'user_message': messages,'message_date': dates})
```

```python
df['message_date']=pd.to_datetime(df['message_date'], format="%d/%m/%y, %I:%M%p - ", dayfirst=True)
```

```python
df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
df
```

```python
# separate users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split(r'([\w\W]+?):\s',message)#return list of name and message
    if entry[1:]: # user name #entry[1:] is a slice of list entry #entry[1:] all elements of list from index 1
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notifications')
        messages.append(entry[0])

df['user']=users
df['message']=messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
df['year']=df['date'].dt.year
```

```python
df['month']=df['date'].dt.month_name()
```

```python
df['day']=df['date'].dt.day
```

```python
df['hour']=df['date'].dt.hour
```

```python
df['minute']=df['date'].dt.minute
```
